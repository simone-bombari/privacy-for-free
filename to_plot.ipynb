```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from utils import import_in_df, import_in_df_copy

sns.set_theme(style="darkgrid")
```

```python
dfs = import_in_df_copy('./data_points/synthetic/', 'ntk_rf')
dfs = dfs.loc[dfs['fmap'] == 'ntk']

dfmnist = import_in_df('./data_points/MNIST/', 'ntk')
dfcifar = import_in_df('./data_points/CIFAR-10/', 'ntk')

dfmnist = dfmnist.loc[dfmnist['N'] <= 5000]
dfcifar = dfcifar.loc[dfcifar['N'] <= 1000]
```

```python
fig, ax = plt.subplots(1, 4, figsize=(11, 3.2))

new_palette = ['#c44e52', 'black']

sns.lineplot(x='N', y='score_test', hue='alpha_min_1', style='phi', data=dfs, ax=ax[0], palette=new_palette)
sns.lineplot(x='N', y='score_attack', hue='alpha_min_1', style='phi', data=dfs, ax=ax[1], palette=new_palette)

sns.lineplot(x='N', y='score_attack', data=dfmnist, ax=ax[2], label='Attack accuracy', color=new_palette[1])
sns.lineplot(x='N', y='score_test', data=dfmnist, ax=ax[2], label= 'Test accuracy', color=new_palette[0])

sns.lineplot(x='N', y='score_attack', data=dfcifar, ax=ax[3], label='Attack', color=new_palette[1])
sns.lineplot(x='N', y='score_test', data=dfcifar, ax=ax[3], label= 'Test', color=new_palette[0])

for i in range(1, 4):
    ax[i].set_ylim(0.48, 1.02)
    ax[i].yaxis.set_ticklabels([])

handles, _ = ax[0].get_legend_handles_labels()
new_handles = [handles[1], handles[2], handles[4], handles[5]]
new_labels = [r'$\phi_2, \alpha=1/2$', r'$\phi_2, \alpha=1/4$', r'$\phi_4, \alpha=1/2$', r'$\phi_4, \alpha=1/4$']
new_handles[2].set_linestyle('--')
new_handles[2].set_color('#c44e52')

ax[0].legend(new_handles, new_labels, loc='upper left')
ax[1].legend().remove()
ax[3].legend().remove()

ax[0].set_ylim(0.48, 1.02)
ax[0].set_ylabel('')
ax[0].set_xlabel(r'$N$')
ax[0].set_xscale('log')
ax[0].set_title('Synthetic - Test accuracy')

ax[1].set_ylabel('')
ax[1].set_xlabel(r'$N$')
ax[1].set_xscale('log')
ax[1].set_title('Synthetic - Attack accuracy')

ax[2].set_ylabel('')
ax[2].set_xlabel(r'$N$')
ax[2].set_xscale('log')
ax[2].set_title('MNIST (1 and 7)')

ax[3].set_ylabel('')
ax[3].set_xlabel(r'$N$')
ax[3].set_xscale('log')
ax[3].set_title('CIFAR-10 (cats and ships)')

fig.tight_layout()
fig.savefig('NTK.pdf')
```

```python
dataset = 'CIFAR-10'
width_noise = 8

num_classes = 10 if dataset == 'CIFAR-10' else 100 if dataset == 'CIFAR-100' else 10
in_channels = 3 if dataset == 'CIFAR-10' else 3  # Assuming CIFAR-10 uses RGB images
img_size = 32 + 2 * width_noise  # Adjusted input image size

# Parameters in layer1
conv1_params = (32 * in_channels * 3 * 3) + 32  # Convolutional layer parameters
layernorm1_params = 32 * 3 * img_size * img_size  # LayerNorm parameters
# Parameters in layer2
conv2_params = (64 * 32 * 3 * 3) + 64  # Convolutional layer parameters
layernorm2_params = 64 * 3 * (img_size // 2) * (img_size // 2)  # LayerNorm parameters
# Parameters in layer3
conv3_params = (128 * 64 * 3 * 3) + 128  # Convolutional layer parameters
layernorm3_params = 128 * 3 * (img_size // 4) * (img_size // 4)  # LayerNorm parameters
# Parameters in fc1
fc1_params = (128 * (img_size // 8) ** 2 * 1024) + 1024  # Linear layer parameters
# Parameters in fc2
fc2_params = (1024 * num_classes) + num_classes  # Linear layer parameters

# Total parameters
total_params = conv1_params + layernorm1_params + conv2_params + layernorm2_params + \
               conv3_params + layernorm3_params + fc1_params + fc2_params

print("Total number of parameters in the BigConvNet model:", total_params)
```

```python
import torch.nn as nn

dataset = 'CIFAR-10'
noise_width = 8

num_classes = 10 if dataset == 'CIFAR-10' else 100 if dataset == 'CIFAR-100' else 10
in_channels = 1 if dataset == 'MNIST' else 3 if dataset == 'CIFAR-10' else 3
img_size = 28 + 2 * noise_width if dataset == 'MNIST' else 32 + 2 * noise_width if dataset == 'CIFAR-10' else 32

# Parameters in conv_layers
conv1_params = (16 * in_channels * 3 * 3) + 16  # Convolutional layer parameters
conv2_params = (32 * 16 * 3 * 3) + 32  # Convolutional layer parameters

# Parameters in fc_layers
fc1_params = (32 * (img_size // 4) * (img_size // 4) * 256) + 256  # Linear layer parameters
fc2_params = (256 * num_classes) + num_classes  # Linear layer parameters

# Total parameters
total_params = conv1_params + conv2_params + fc1_params + fc2_params

print("Total number of parameters in the SmallConvNet model:", total_params)
```
