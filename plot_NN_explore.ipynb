```python
import os
import json
import numpy as np
import pandas as pd
```

```python
def load_json_files_to_dataframe(directory):
    # Initialize an empty list to store data
    data_list = []
    
    # Loop over all files in the specified directory
    for filename in os.listdir(directory):
        if filename.endswith('.json'):  # Check if the file is a JSON file
            file_path = os.path.join(directory, filename)
            
            # Open and read the JSON file
            with open(file_path, 'r') as file:
                data = json.load(file)
                data_list.append(data)
    
    # Convert the list of dictionaries to a DataFrame
    df = pd.DataFrame(data_list)
    
    return df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='darkgrid')
```

```python
def explode_dataframe(df):
    # Explode the lists into separate rows
    df_exploded = df.explode(['train_losses', 'test_losses', 'test_accuracies', 'grad_norms_param1', 'grad_norms_param2'])
    
    # Create an 'epoch' column based on the index within each exploded group
    df_exploded['epoch'] = df_exploded.groupby(['width', 'lr']).cumcount()
    
    return df_exploded


def plot_results(df, subplot='lr', hue='width', to_plot='train_losses'):
    # Validate input arguments
    if subplot not in ['lr', 'width']:
        raise ValueError("subplot must be either 'lr' or 'width'")
    if hue not in ['lr', 'width']:
        raise ValueError("hue must be either 'lr' or 'width'")
    if to_plot not in ['train_losses', 'test_losses', 'test_accuracies', 'grad_norms_param1', 'grad_norms_param2']:
        raise ValueError("to_plot must be one of 'train_losses', 'test_losses', 'test_accuracies', 'grad_norms_param1', 'grad_norms_param2']")
    
    # Explode the dataframe to handle lists
    df_exploded = explode_dataframe(df)
    
    # Set up the FacetGrid for subplots
    g = sns.FacetGrid(df_exploded, col=subplot, hue=hue, col_wrap=3, height=4, aspect=1.5)
    
    # Map the data to the grid
    g = g.map(sns.lineplot, 'epoch', to_plot).add_legend()
    
    # Set y-axis limits based on the 'to_plot' argument
    if to_plot in ['train_losses', 'test_losses']:
        g.set(ylim=(0, 5))
    elif to_plot == 'test_accuracies':
        g.set(ylim=(0, 100))
    elif to_plot == 'grad_norms_param1':
        g.set(ylim=(0, 20))
    elif to_plot == 'grad_norms_param2':
        g.set(ylim=(0, 40))

    # Adjust the labels and titles
    g.set_axis_labels("Epochs", to_plot.replace('_', ' ').capitalize())
    g.set_titles(col_template="{col_name}", row_template="{row_name}")
    
    plt.subplots_adjust(top=0.9)
    g.fig.suptitle(f'{to_plot.replace("_", " ").capitalize()} over Epochs by {hue.capitalize()} and {subplot.capitalize()}', fontsize=16)
    
    plt.show()
```

```python
def plot_final_metric(df, x='width', hue='lr', to_plot='test_accuracies', norm_param=None, epoch_range=None):
    """
    Plots the final test accuracy, average gradient norm, or overall norm across epochs with log scale for width and learning rate.
    
    Parameters:
    - df: DataFrame containing the results.
    - x: The variable to plot on the x-axis ('width' or 'lr').
    - hue: The variable to use for color grouping ('width' or 'lr').
    - to_plot: The metric to plot ('test_accuracies', 'norm', or 'overall_norm').
    - norm_param: Specifies which gradient norm to plot if `to_plot` is 'norm' ('grad_norms_param1' or 'grad_norms_param2').
    - epoch_range: Tuple specifying the range of epochs to average the gradient norm over (e.g., (0, 100)).
    """
    # Validate input arguments
    if x not in ['lr', 'width']:
        raise ValueError("x must be either 'lr' or 'width'")
    if hue not in ['lr', 'width']:
        raise ValueError("hue must be either 'lr' or 'width'")
    if to_plot not in ['test_accuracies', 'norm', 'overall_norm']:
        raise ValueError("to_plot must be 'test_accuracies', 'norm', or 'overall_norm'")
    if to_plot == 'norm' and norm_param not in ['grad_norms_param1', 'grad_norms_param2']:
        raise ValueError("norm_param must be either 'grad_norms_param1' or 'grad_norms_param2' when to_plot is 'norm'")
    
    # Explode the dataframe to handle lists
    df_exploded = explode_dataframe(df)
    
    # Create log-transformed columns
    df_exploded['log_lr'] = np.log10(df_exploded['lr'])
    df_exploded['log_width'] = np.log10(df_exploded['width'])
    
    # Calculate the overall norm if required
    if to_plot == 'overall_norm':
        df_exploded['overall_norm'] = np.sqrt(df_exploded['grad_norms_param1']**2 + df_exploded['grad_norms_param2']**2)
    
    # Prepare data based on the metric to plot
    if to_plot == 'test_accuracies':
        # Get the final test accuracy (last epoch)
        df_exploded['final_test_accuracy'] = df_exploded.groupby(['log_width', 'log_lr'])['test_accuracies'].transform('last')
        plot_data = df_exploded.drop_duplicates(subset=['log_width', 'log_lr'])
        y_label = 'Final Test Accuracy'
    elif to_plot == 'norm':
        if epoch_range:
            start_epoch, end_epoch = epoch_range
            plot_data = df_exploded[(df_exploded['epoch'] >= start_epoch) & (df_exploded['epoch'] <= end_epoch)]
        else:
            plot_data = df_exploded
        plot_data = plot_data.copy()  # Avoid the SettingWithCopyWarning
        plot_data[f'avg_{norm_param}'] = plot_data.groupby(['log_width', 'log_lr'])[norm_param].transform('mean')
        plot_data = plot_data.drop_duplicates(subset=['log_width', 'log_lr'])
        y_label = f'Avg {norm_param.replace("_", " ").capitalize()}'
    elif to_plot == 'overall_norm':
        if epoch_range:
            start_epoch, end_epoch = epoch_range
            plot_data = df_exploded[(df_exploded['epoch'] >= start_epoch) & (df_exploded['epoch'] <= end_epoch)]
        else:
            plot_data = df_exploded
        plot_data = plot_data.copy()  # Avoid the SettingWithCopyWarning
        plot_data['avg_overall_norm'] = plot_data.groupby(['log_width', 'log_lr'])['overall_norm'].transform('mean')
        plot_data = plot_data.drop_duplicates(subset=['log_width', 'log_lr'])
        y_label = 'Avg Overall Norm'
    
    # Define x and hue based on log columns
    x_col = 'log_width' if x == 'width' else 'log_lr'
    hue_col = 'log_lr' if hue == 'lr' else 'log_width'
    
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Plot with lineplot
    sns.lineplot(x=x_col, y=f'final_test_accuracy' if to_plot == 'test_accuracies' else f'avg_{norm_param}' if to_plot == 'norm' else 'avg_overall_norm',
                 hue=hue_col, data=plot_data, palette='viridis', linewidth=2.5, ax=ax)
    
    # Set labels
    plt.xlabel(f'Log {x.capitalize()}', fontsize=14)
    plt.ylabel(y_label, fontsize=14)
    plt.title(f'{y_label} by Log {x.capitalize()} and Log {hue.capitalize()}', fontsize=16)
    
    plt.show()
```

```python
def plot_time_to_threshold(df, threshold=90, x='width', hue='lr', y_transform=1):
    """
    Plots the first epoch where the test accuracy exceeds the given threshold as a function of width and learning rate.
    
    Parameters:
    - df: DataFrame containing the results.
    - threshold: The test accuracy threshold to find the first epoch where it's exceeded.
    - x: The variable to plot on the x-axis ('width' or 'lr').
    - hue: The variable to use for color grouping ('width' or 'lr').
    - y_transform: The power to which the y variable (time to threshold) is raised (e.g., 1/2 for square root).
    """
    time_to_threshold = []

    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        accuracies = row['test_accuracies']
        
        # Find the first epoch where the accuracy exceeds the threshold
        try:
            first_epoch = next(i for i, acc in enumerate(accuracies) if acc > threshold)
            time_to_threshold.append(first_epoch)
        except StopIteration:
            # If the threshold is never reached, we append a NaN value
            time_to_threshold.append(np.nan)
    
    # Add the time to threshold to the DataFrame
    df['time_to_threshold'] = time_to_threshold
    
    # Drop rows where the threshold was never reached
    plot_data = df.dropna(subset=['time_to_threshold']).copy()
    
    # Apply log transformation to the hue variable
    plot_data['log_hue'] = np.log10(plot_data[hue])
    
    # Apply the custom transformation to the y variable (time to threshold)
    plot_data['transformed_time_to_threshold'] = plot_data['time_to_threshold'] ** y_transform
    
    # Define x and hue based on original columns
    x_col = x
    hue_col = 'log_hue'
    
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Plot with lineplot
    sns.lineplot(x=x_col, y='transformed_time_to_threshold', hue=hue_col, data=plot_data, palette='viridis', linewidth=2.5, ax=ax)
    
    # Set labels
    plt.xlabel(f'{x.capitalize()}', fontsize=14)
    plt.ylabel(f'Time to {threshold}% Accuracy^{y_transform}', fontsize=14)
    plt.title(f'Transformed Time to {threshold}% Accuracy by {x.capitalize()} and Log {hue.capitalize()}', fontsize=16)
    
    plt.show()
```

```python
plot_time_to_threshold(df, threshold=60, x='width', hue='lr')
```

```python
import os
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


def load_data(folder_path):
    # List all json files in the directory
    json_files = [f for f in os.listdir(folder_path) if f.endswith('.json')]
    
    # Prepare a list to hold all rows before creating DataFrame
    rows = []
    
    # Process each file
    for filename in json_files:
        with open(os.path.join(folder_path, filename), 'r') as file:
            data = json.load(file)
        
        # Extract mode from the filename
        if 'full' in filename:
            mode = 'full'
        elif 'no_noise' in filename:
            mode = 'no_noise'
        elif 'no_clip' in filename and 'no_clip_with_noise' not in filename:
            mode = 'no_clip'
        elif 'no_clip_with_noise' in filename:
            mode = 'no_clip_with_noise'
        else:
            mode = 'unknown'  # Default case if no known mode is found
        
        # Append mode to the data
        data['mode'] = mode
        
        # Append this entry to the list
        rows.append(data)
    
    # Create DataFrame
    df = pd.DataFrame(rows)
    
    return df

# Usage
folder_path = './NN/___'  # Change this to your folder path
df = load_data(folder_path)

df
```

```python
# import seaborn as sns
# import matplotlib.pyplot as plt
# import numpy as np
# import pandas as pd

def plot_combined_metrics(df, metric):
    sns.set_theme(style="darkgrid")

    # Extract unique values for plotting grid
    modes = df['mode'].unique()
    Ts = sorted(df['T'].unique())  # Sort T values to ensure the plots are in ascending order

    # Prepare the figure grid
    fig, axes = plt.subplots(nrows=len(modes), ncols=len(Ts), figsize=(15, 15), sharey=True)
    if len(modes) * len(Ts) == 1:
        axes = np.array([axes])

    # Initialize a variable to track whether the legend has been created
    legend_created = False

    for i, mode in enumerate(modes):
        for j, T in enumerate(Ts):
            ax = axes[i, j] if axes.ndim > 1 else axes[max(i, j)]
            # Filter data for current mode and T value
            plot_data = df[(df['mode'] == mode) & (df['T'] == T)]

            # Make sure there's data to plot
            if not plot_data.empty:
                # Unpack the list of metrics into a new DataFrame for easy plotting
                epoch_data = []
                max_epoch = 0  # Variable to find the maximum epoch for setting x-axis limits
                for index, row in plot_data.iterrows():
                    epochs = np.arange(len(row[metric]))
                    max_epoch = max(max_epoch, epochs[-1])  # Update the max epoch encountered
                    for epoch, value in zip(epochs, row[metric]):
                        epoch_data.append({
                            'Epoch': epoch,
                            metric: value,
                            'lr': row['lr'],
                            'batch_size': row['batch_size']
                        })
                expanded_df = pd.DataFrame(epoch_data)

                # Plot using seaborn lineplot
                line = sns.lineplot(data=expanded_df, x='Epoch', y=metric, hue='lr', style='batch_size', 
                                    ax=ax) #, markers=True, dashes=False)

                # Create a single legend in the first subplot and then avoid creating more legends
                if not legend_created:
                    ax.legend(title='Configurations')
                    legend_created = True
                else:
                    ax.legend([],[], frameon=False)

            ax.set_title(f'Mode: {mode}, T: {T}')
            ax.set_xlabel('Epoch')
            ax.set_xlim(0, max_epoch + 1)  # Adjust x-axis limits to max epoch + 1 for buffer
            ax.set_ylabel(metric)

            # ax.set_xscale('log')
            ax.set_yscale('log')

    plt.tight_layout()
    plt.subplots_adjust(top=0.92)  # Adjust the top margin to fit the super title without overlapping
    fig.suptitle(f'Combined Metrics for {metric}', fontsize=16)
    plt.show()
```
